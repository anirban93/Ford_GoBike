```python
data=df.dropna()
```

```python
df.info()
```

```python
data['duration_hour']=data['duration_sec'] / 3600
```

```python
data.head()
```

```python
data.start_time= pd.to_datetime(data.start_time)
data.end_time= pd.to_datetime(data.end_time)
```

```python
df.info()
```

```python
df.head()
```

```python
plt.hist(data.start_time.dt.hour,color='c',label='start_time')

plt.xlabel('Hour')
plt.ylabel('Count')

plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.hist(data.start_time.dt.day_name(),color='c',width=0.5,label='start_time')

plt.xlabel('week days')
plt.ylabel('Count')

plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data.start_time.dt.date,color='c',label='start_time')

plt.xlabel('Date')
plt.ylabel('Count')

plt.legend()
plt.show()
```

```python
grouped_birth_year=data.groupby('member_birth_year')['duration_hour']
birth_year_age=grouped_birth_year.agg(['mean','count','max'])
```
